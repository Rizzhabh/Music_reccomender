```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn import tree
import matplotlib.pyplot as plt

music_data = pd.read_csv(r'C:\Users\rizzh\Downloads\music.csv')
input_set = music_data.drop(columns = ["genre"])
output_set = music_data['genre']

x_train , x_test , y_train , y_test = train_test_split(input_set, output_set, test_size= 0.2)


model = DecisionTreeClassifier()
model.fit(x_train,y_train)
predictions = model.predict(x_test)

accuracy = accuracy_score(y_test, predictions)
print(f"Model accuracy: {accuracy}")

plt.figure(figsize=(15,10))
tree.plot_tree(model,
               feature_names= ['age', 'gender'] ,
               class_names=sorted(output_set.unique()),
               filled=True,
               rounded=True,
               fontsize=10)
plt.title("Music Genre Decesion Tree Based on Age and gender")
plt.tight_layout()
plt.show()



```
